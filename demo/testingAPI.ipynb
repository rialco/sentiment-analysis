```python
import requests
import pandas as pd

subject = 'ingrodolfohdez'
url = 'http://localhost:3000/sentiment/analysis/' + subject

response = requests.get(url)
data = response.json()


df = pd.DataFrame(data)
cols = ['city', 'sentiment']
df = df[cols[:]]
df['sentiment'] = df['sentiment'].astype(int)
global_df = df.groupby('sentiment').size()
global_df.plot(kind='bar')

city_df = df.groupby(['city', 'sentiment']).size().unstack(fill_value=0)
city_df.plot(kind='bar', width=1)
```
